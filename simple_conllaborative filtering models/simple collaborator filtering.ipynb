```python
from data_preprocessing import data_preprocessing
import numpy as np
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
```

```python
def calculate_distance(vector_a, vector_b):
    tempt = np.subtract(vector_a,vector_b)
    num_of_users = np.square(tempt)
    individual_difference = np.average(num_of_users)
    individual_difference = np.sqrt(individual_difference)
    return individual_difference

def calculate_overall_distance_matrix(data):
    num_of_users = np.shape(data)[0]
    distance_matrix = np.zeros((num_of_users, num_of_users))

    for i in np.arange(num_of_users):
        for j in np.arange(num_of_users):
            first_parameter = data[i]
            second_parameter = data[j]
            distance_matrix[i][j] = calculate_distance(first_parameter, second_parameter)

    return distance_matrix

def find_nearest_neighbours(data, distance_matrix,index):
    a = distance_matrix[index]
    nearest_neighbours = np.sort(distance_matrix[index])[0:100]
    index_list = []
    for i in nearest_neighbours:
        for j in  np.asarray(np.where(a == i)[0]):
            index_list.append( j )
    index_list = np.asarray(index_list)

    result = []

    for i in index_list:
        result.append(data[i])

    return result

def setup_value(data,distance_matrix):
    new_data = np.array([])
    num_of_users = np.shape(data)[0]

    for index in np.arange(num_of_users):
        #print('the %s user calculation starts'% index)
        a_vector = np.asarray(np.copy(data[index]))

        nearest_neighbours = find_nearest_neighbours(data, distance_matrix,index)
        counter = np.shape(a_vector)[0]

        for i in np.arange(counter):
            if a_vector[i]==0:

                for a_neighbour in nearest_neighbours:
                    if a_neighbour[i] !=0:
                        a_vector[i] = a_neighbour[i]
                        break

        a_vector = np.reshape(a_vector,(636,1))
        new_data = np.append(new_data,a_vector)
        new_data = np.reshape(new_data,(-1,636))
        # print((a_vector))
        # data[index] = a_vector
    return new_data
```

```python
distance_matrix = calculate_overall_distance_matrix(data)
data = setup_value(data,distance_matrix)

print(data)
print('execution finishes here')
```
